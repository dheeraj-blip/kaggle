```python
unique_values = df.apply(lambda x: x.unique())

# Display unique values
for column, unique in unique_values.items():
    print(f"Unique values in column '{column}': {unique}")
```

```python

# Define the mapping
mapping = {
    'mainroad': {'yes': 1, 'no': 0},
    'guestroom': {'yes': 1, 'no': 0},
    'basement': {'yes': 1, 'no': 0},
    'hotwaterheating': {'yes': 1, 'no': 0},
    'airconditioning': {'yes': 1, 'no': 0},
    'prefarea': {'yes': 1, 'no': 0},
    'furnishingstatus': {
        'furnished': 1,
        'semi-furnished': 0.5,
        'unfurnished': 0
    }
}

# Replace values according to the mapping
for column, value_map in mapping.items():
    df[column] = df[column].replace(value_map)

df
```

```python
correlation_matrix = df.corr()


import seaborn as sns
import matplotlib.pyplot as plt

# Set the size of the plot
plt.figure(figsize=(10, 8))

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True, cbar_kws={"shrink": .8})

# Display the heatmap
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics import mean_squared_error, r2_score



# Prepare your data
X = df.drop(columns=['price'])  # Features
y = df['price']                  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Initialize Kernel Ridge Regressor
model = KernelRidge(kernel='rbf', alpha=1.0)  # You can adjust the kernel and alpha as needed

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Root Mean Squared Error:", mse**0.5)

```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Prepare your data
X = df.drop(columns=['price'])  # Features
y = df['price']                  # Target variable

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Initialize Linear Regression model
model = LinearRegression()  # No hyperparameters to set in basic Linear Regression

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Root Mean Squared Error:", mse**0.5)

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import flaml

# Prepare your dataset
# Assuming `df` is your dataset and 'price' is the target variable
X = df.drop(columns=['price'])  # Features
y = df['price']                 # Target variable

# Split data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# Initialize and run AutoML
automl = flaml.AutoML()

# FLAML settings for regression
settings = {
    "time_budget": 60,  # Total time in seconds
    "metric": 'r2',  # Metric to optimize ('r2', 'rmse', etc.)
    "task": 'regression',  # Task type (can be 'classification' or 'regression')
    "log_file_name": 'flaml.log',  # Log file
}

# Train the model
automl.fit(X_train, y_train, **settings)

# Make predictions
y_pred = automl.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

# Print the best model found by FLAML
print("Best model:", automl.best_estimator)
```
